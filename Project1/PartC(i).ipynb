```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action='ignore', category=(FutureWarning, UserWarning, DeprecationWarning))
```

```python
merge_sort_i_df = pd.read_csv('./data/merge_sort_with_insertion_sort_diff_n.csv')
# merge_sort_df['CompareCountAvg'] = (merge_sort_df['CompareCount1'] + merge_sort_df['CompareCount2'] + merge_sort_df['CompareCount3'])/3
#merge_sort_df['timeAvg'] = (merge_sort_df['time1'] + merge_sort_df['time2'] + merge_sort_df['time3'])/3
merge_sort_i_df.head(50)
```

```python

# line plot for key comparison
f, ax = plt.subplots(1, 2, figsize=(15, 5))

sb.lineplot(
    data=merge_sort_i_df,
    y='compareCount',
    x='n',
    ax = ax[0]
)
ax[0].set_title('Key Comparison') #set title for first subplot

# line plot for time
sb.lineplot(
    data=merge_sort_i_df,
    y='time',
    x='n',
    ax=ax[1] #use the second subplot
)
ax[1].set_title('Time')  #set title for second subplot

#tight layout to prevent overlapping
plt.tight_layout()

plt.show()
```
