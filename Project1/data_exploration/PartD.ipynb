```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("./data/compare_hybrid_with_original.csv", index_col=0)
```

```python
data.head()
```

```python
hybrid_compare = data["hybrid_compare"]
original_compare = data["original_compare"]
hybrid_time = data["hybrid_time"]
original_time = data["original_time"]
```

```python
plt.figure(figsize=(10, 2))
plt.title("Key Comparison")
sns.stripplot(data={"Hybrid": hybrid_compare, "Original": original_compare}, orient="h")
```

```python
plt.figure(figsize=(10, 2))
plt.title("Time (ms)")
sns.stripplot(data={"Hybrid": hybrid_time, "Original": original_time}, orient="h")
```

```python
plt.figure(figsize=(10, 2))
plt.title("Time (ms)")
sns.boxplot(data=pd.DataFrame({"Hybrid": hybrid_time, "Original": original_time}), orient="h")
```

```python
inc = 0
for i in range(1, 101):
    inc += 1 - hybrid_time[i] / original_time[i]
print("Average time improvement: {:.2f}%".format(inc / 100 * 100)) 
```
